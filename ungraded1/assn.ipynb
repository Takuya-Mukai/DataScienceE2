## Q1: difference between analytical and numerical probability
- Analytical probability of getting exact results by using probability theory
- Numerical probability of getting exact results by using simulations

## Q2: benefits of numerical probability calculations
- Numerical probability calculations can handle complex systems and scenarios where analytical solutions are difficult or impossible to derive.

## Q3: is this a probability distribution?
- No, it's not, because the sine wave oscillates between positive and negative values, and a probability distribution must be non-negative everywhere.

```python
# Task2: numerical mean and variance
a = np.random.normal(size=1000)

mean = a.mean()
sigma = np.array([(i - a.mean())**2 for i in a]).mean()
print("Mean:", mean)
print("Variance:", sigma)
```
